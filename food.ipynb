# 6. DATA CLEANING

-In this step we handle missing values , remove duplicates and prepare the dataset for analysis. 



## Check missing values
checking missing values in the dataset that may affect analysis

```python
#missing values
df.isnull().sum()
```

## Handle missing values
There are missing values in several columns.

-numerical columns

-categorical columns

check percentage of missing values

```python
df.isnull().sum()/len(df)*100
```

Fill missing values

```python
#fill empty cells for GPA
df.fillna({"GPA":3.5},inplace=True)
```

```python
#fill empty cells for calories day
df.fillna({"calories_day":3},inplace=True)
```

```python
#fill empty cells for type_sports
df.fillna({"type_sports":"Football",},inplace=True)
```

```python
#fill empty cells for weight
df.fillna({"weight":100},inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert numeric codes to labels
df["Gender"] = df["Gender"].map({1: "Male", 2: "Female"})
```

Check duplicates

```python
#check duplicates
df.duplicated().sum()
```

```python
#cleaned dataset
df.to_csv("cleaned_food_coded_dataset.csv", index=False)
```

# 7. EXPLORATORY DATA ANALYSIS

In this sector,we explore trends and relationships between variables related to students' eating habits and lifestyle.

## I. UNIVARIATE ANALYSIS
-Analyzing individual variables to understand their characteristics and distibution.

### * GPA Distribution

-The plot shows how GPA values are distributed among the students.

```python
#convert to numeric
df["GPA"] = pd.to_numeric(df["GPA"], errors='coerce')
#plot a histogram
sns.histplot(df['GPA'],bins=10,kde=True)
#title 
plt.title("GPA DISTRIBUTION")
#label
plt.xlabel("GPA")
plt.ylabel("frequency")
#display
plt.show
```

# Findings

-The GPA distribution shows that most students have high performance with GPA between 3.0 to 3.8.

-There are few students with low GPA points.

-This suggest overall strong academic performance among students

### * Gender distribution
-counts how many students fall into each gender category 

```python
#plot a countplot
sns.countplot(x='Gender',data=df)
#title
plt.title("GENDER DISTRIBUTION")
plt.show()
```

## Findings

-The plot shows that most students in the college are male with fewer female.

-This indicates that the dataset is slightly skewed towards male students, which may influence the overall analysis.

### *Fruits consumption


```python
#create histogram
sns.histplot(df['fruit_day'], bins=7,kde=True)
#title
plt.title("Fruit Consumption per Day")
#display
plt.show()
```

## Findings

-The graph shows how often students consume fruits daily. 

-Fruit consumption varies among students.

-Most students consume fruits at a moderate level, with fewer students having very high or very low intake.

### * Vegetables consumption


```python
#create histogram
sns.histplot(df['veggies_day'], bins=7,kde=True)
#title
plt.title("Vegetabes consumption per day")
#display
plt.show()
```

## Findings

-The graph shows how often students consume vegetables daily. 

-vegetables consumption varies among students.

-Some students consume vegetables regularly, while others have lower intake, suggesting inconsistent healthy eating habits.

# II BIVARIATE ANALYSIS

-Examines analyis between two variables

### * GPA VS Gender

Compares GPA distribution between genders

```python
#create boxplot
sns.boxplot(x='Gender',y="GPA",data=df)
#title
plt.title("GPA VS GENDER")
#Display
plt.show()
```

## Findings

-The boxplot compares the distribution of GPA between male and female students. 

-The median GPA for the male students is 3.50 while that of femmale students is around 3.40.

-The male students have slightly higher GPA than the female. 

-There is no major difference in GPA between male and female students.

### * Breakfast frequency 

Counts the number of times the students take breakfast.

```python
#create countplot
sns.countplot(x='Gender', data=df)
#title
plt.title("BREAKFAST FREQUENCY")
#display
plt.show()
```

### Findings

-The graph shows that the male students have breakfast more than the female students.

-This suggests a difference in eating habits between genders, where males are more likely to maintain regular breakfast consumption compared to females.

## * Coffee frequency

Displays the number of times each gender take coffee.

```python
#create countplot
sns.countplot(x='Gender', data=df)
#title
plt.title("COFFEE FREQUENCY")
#display
plt.show()
```

## Findings

--The graph shows that the male students have coffee more frequently than the female students.

-This suggests a difference in eating habits between genders, where males are more likely to maintain regular coffee consumption compared to females.

### *Eating out frequency

Displays the number of times the students eat out

```python
#create countplot
sns.countplot(x='Gender', data=df)
#title
plt.title("EATING OUT FREQUENCY")
#display
plt.show()
```

### * Cooking frequency
Displays the number of times the students cook.

```python
#create countplot
sns.countplot(x='cook',hue='Gender', data=df)
#title
plt.title("COOKING FREQUENCY")
#display
plt.show()
```

## Findings

-The graph shows differences in cooking frequency between male and female students.

-Male students are more concentrated in the lower and moderate cooking levels (1–3), while female students appear more in the higher cooking levels (4–5).

-This suggests that female students tend to cook more frequently than male students.

## III MULTIVARIATE ANALYSIS

Studies three or more variables together.

### *Correlation heatmap

```python
# Set figure size 
plt.figure(figsize=(10,6))
#select numeric columns 
numeric_df = df.select_dtypes(include=['number'])
# Create heatmap to show relationships between numerical variables
sns.heatmap(numeric_df.corr(), annot=True)
# title
plt.title("Correlation Between Variables")

# Display
plt.show()
```

## Findings

-The heatmap shows how different variables are related to each other.

-Most values are close to zero,which means there is little relationship between variables.This shows that eating habits of students are mostly independent.
